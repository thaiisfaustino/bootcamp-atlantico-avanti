## Visualização dos dados

### Dados qualitativos

```python
# Analisando a variável Sex
print("Variáveis:", df_tips['sex'].unique())
```

```python
fig = sns.displot(data = df_tips, x = "sex")
plt.show(fig)
```

```python
# Analisando a variável Day
print("Variáveis:", df_tips['day'].unique())
```

```python
fig = sns.displot(data = df_tips, x = "day")
plt.show(fig)
```

```python
print("Variáveis:", df_tips['time'].unique())
```

### Dados quantitativos

```python
print("Media:", df_tips['total_bill'].mean().round(2))
print("Mediana:", df_tips['total_bill'].median())
print("Moda:", df_tips['total_bill'].mode())
print("Amplitude:", df_tips['total_bill'].max() - df_tips['total_bill'].min())
print("Variância:", df_tips['total_bill'].var().round(2))
print("Desvio Padrão:", df_tips['total_bill'].std().round(2))
```

```python
print("Media:", df_tips['tip'].mean().round(2))
print("Mediana:", df_tips['tip'].median())
print("Moda:", df_tips['tip'].mode())
print("Amplitude:", df_tips['tip'].max() - df_tips['tip'].min())
print("Variância:", df_tips['tip'].var().round(2))
print("Desvio Padrão:", df_tips['tip'].std().round(2))
```

```python
print("Media:", df_tips['size'].mean().round(2))
print("Mediana:", df_tips['size'].median())
print("Moda:", df_tips['size'].mode())
print("Amplitude:", df_tips['size'].max() - df_tips['size'].min())
print("Variância:", df_tips['size'].var().round(2))
print("Desvio Padrão:", df_tips['size'].std().round(2))
```
